```python
import os
import pandas as pd

import pandas as pd

file_path = '/content/dataset_ideology.csv'
df = pd.read_csv(file_path)

print(df.head())
```

```python
ideology_stats = df.groupby('predicted_ideology').agg({
    'likes': 'mean',
    'retweets': 'mean'
}).reset_index()

print(ideology_stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x='predicted_ideology', y='likes', data=ideology_stats, palette='viridis', ci=None)
plt.title('Mean Likes by Ideology')
plt.ylabel('Mean Likes')
plt.xlabel('Ideology')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='predicted_ideology', y='retweets', data=ideology_stats, palette='viridis', ci=None)
plt.title('Mean Retweets by Ideology')
plt.ylabel('Mean Retweets')
plt.xlabel('Ideology')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
